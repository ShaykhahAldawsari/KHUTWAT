```python
df = df.drop('Patient Number', axis = 1)
df.head()
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
df['Mood Swing'] = label.fit_transform(df['Mood Swing'])
df['Sleep dissorder'] = label.fit_transform(df['Sleep dissorder'])
df['Suicidal thoughts'] = label.fit_transform(df['Suicidal thoughts'])
df['Anorxia'] = label.fit_transform(df['Anorxia'])
df['Authority Respect'] = label.fit_transform(df['Authority Respect'])
df['Try-Explanation'] = label.fit_transform(df['Try-Explanation'])
df['Aggressive Response'] = label.fit_transform(df['Aggressive Response'])
df['Ignore & Move-On'] = label.fit_transform(df['Ignore & Move-On'])
df['Nervous Break-down'] = label.fit_transform(df['Nervous Break-down'])
df['Admit Mistakes'] = label.fit_transform(df['Admit Mistakes'])
df['Overthinking'] = label.fit_transform(df['Overthinking'])
```

```python
# replace the frequency values with numbers
dict = {'Seldom': 1, 'Sometimes': 2, 'Usually': 3, 'Most-Often': 4}
df['Sadness'] = df['Sadness'].replace(dict).astype(int)
df['Euphoric'] = df['Euphoric'].replace(dict).astype(int)
df['Exhausted'] = df['Exhausted'].replace(dict).astype(int)
df['Sleep dissorder'] = df['Sleep dissorder'].replace(dict).astype(int)
```

```python
df['Expert Diagnose'] = df['Expert Diagnose'].map({'Normal': 0, 'Bipolar Type-1': 1,'Bipolar Type-2': 2,'Depression': 3}).astype(int)
```

```python
dict = {'3 From 10': 3, '4 From 10': 4, '6 From 10': 6, '5 From 10': 5, '7 From 10': 7, '8 From 10': 8,
 '9 From 10': 9, '2 From 10': 2, '1 From 10': 1}
df['Sexual Activity'] = df['Sexual Activity'].replace(dict).astype(int)
df['Concentration'] = df['Concentration'].replace(dict).astype(int)
df['Optimisim'] = df['Optimisim'].replace(dict).astype(int)
df.head()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.drop(columns=['Expert Diagnose']).corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation')
plt.show()
```
